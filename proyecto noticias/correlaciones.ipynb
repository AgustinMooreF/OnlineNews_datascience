```python
# Tratamiento de datos
# ==============================================================================
import pandas as pd
import numpy as np


# Gr치ficos
# ==============================================================================
import matplotlib.pyplot as plt
from matplotlib import style

# Preprocesado y an치lisis
# ==============================================================================
import pingouin as pg
# Configuraci칩n matplotlib
# ==============================================================================
plt.style.use('ggplot')
import seaborn as sns
# Configuraci칩n warnings
# ==============================================================================
import warnings
warnings.filterwarnings('ignore')
```

```python
datos = pd.read_csv('OnlineNewsPopularity.csv')
```

```python
#se borran espacios en las keys de cada columna
datos.columns = datos.columns.str.replace(' ', '')
```

```python

# for col in datos.columns:
#   se dropea URL ya que para efectos de este trabajo no sirve
datos.drop(columns=['url'],inplace=True)
datos.keys()
# print(datos[' shares'])
```

```python
fig, ax = plt.subplots(1, 1, figsize=(6,4))
ax.scatter(x=datos.title_subjectivity, y=datos.title_sentiment_polarity, alpha= 1)
ax.set_xlabel('title_subjectivity') 
ax.set_ylabel('title_sentiment_polarity')
```

```python
display(pg.corr(datos['title_subjectivity'], datos['title_sentiment_polarity'], method='pearson'))
display(pg.corr(datos['title_subjectivity'], datos['title_sentiment_polarity'], method='spearman'))
```
